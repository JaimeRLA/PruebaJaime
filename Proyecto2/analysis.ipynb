# Análisis de cada Carrera

Ejemplo de simulación de una estación de servicio

Para realizar un análisis de los datos del DataFrame resultante, puedes realizar varias operaciones utilizando las funciones y métodos proporcionados por la biblioteca Pandas en Python. Aquí hay algunas sugerencias sobre cómo podrías analizar los datos:

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('./simulation_statistics.csv')
df2= pd.read_csv('./race_classification.csv')
```

```python
# Gráfico de barras para comparar la cantidad de vehículos de diferentes tipos
df['Fuel Type'].value_counts().plot(kind='bar')
plt.title('Número de vehículos por tipo de combustible')
plt.xlabel('Tipo de combustible')
plt.ylabel('Cantidad de vehículos')
plt.show()
```

## Cálculo de estadísticas resumidas

Calcula estadísticas resumidas, como la media, la mediana, el máximo y el mínimo de ciertas columnas relevantes en tus datos. Por ejemplo, puedes calcular la duración promedio de recarga para cada tipo de combustible.

```python
summary_stats = df.groupby('Fuel Type')['End Refuel Time'].mean().round(2)
print(summary_stats)
```

## Análisis de distribución de datos

Examina la distribución de ciertas variables en tus datos para comprender mejor su comportamiento. Por ejemplo, puedes trazar un histograma de la duración de recarga para todos los vehículos.

```python
plt.hist(df['End Refuel Time'], bins=10)
plt.title('Distribución de la duración de recarga')
plt.xlabel('Duración de recarga')
plt.ylabel('Frecuencia')
plt.show()
```

## Comparación de variables

Realiza comparaciones entre diferentes variables en tus datos para identificar posibles relaciones. Por ejemplo, puedes comparar la duración de recarga con el tiempo de llegada de los vehículos con un diagrama de dispersión para comparar la duración de recarga con el tiempo de llegada.

```python
plt.scatter(df['Arrival Time'], df['End Refuel Time'])
plt.title('Comparación entre tiempo de llegada y duración de recarga')
plt.xlabel('Tiempo de llegada')
plt.ylabel('Duración de recarga')
plt.show()
```

## Análisis de tiempo de espera

Calcula el tiempo de espera promedio de los vehículos en la estación de carga y analiza si hay algún patrón o tendencia en estos tiempos de espera.

```python
df['Wait Time'] = df['Start Refuel Time'] - df['Arrival Time']
average_wait_time = df['Wait Time'].mean().round(2)
print(f'Tiempo de espera promedio: {average_wait_time}')
```

## Identificación de picos de carga

Identifica períodos de tiempo con una alta demanda de carga y analiza qué factores pueden estar contribuyendo a estos picos. Por ejemplo, puedes agrupar los datos por hora del día y calcular el número medio de vehículos que llegan en cada hora.

```python
df['Hour'] = pd.to_datetime(df['Arrival Time']).dt.hour
average_arrivals_per_hour = df.groupby('Hour').size().mean().round(2)
print(f'Número medio de llegadas por hora: {average_arrivals_per_hour}')
```
